```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(columns=['store_id','date','status','revenue'])
```

```python
df['date'] = list(pd.date_range(start='1/1/2018', end='1/30/2018'))*10
df['store_id'] = list(range(1,6))*60
status = []
revenue = []

for i in range(300):
    status.append(np.random.choice(['open','closed','fraud']))
    revenue.append(np.random.randint(0,101))

df['revenue'], df['status'] = revenue, status
```

```python
def frau_percentage(df):
    active = df[df['revenue'] > 0].copy()
    active['is_fraud'] = active['status'] == 'fraud'
    return pd.DataFrame(active.groupby('date')[['is_fraud']].mean()*100).reset_index()
```

```python
frau_percentage(df)
```
